```
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
X = pd.read_csv('ENTRENAMIENTO_ENTRADA.csv').values
Y = pd.read_csv('ENTRENAMIENTO_SALIDA.csv').values
Xtest = pd.read_csv('PRUEBA_ENTRADA.csv').values
Ytest = pd.read_csv('PRUEBA_SALIDA.csv').values
Ytest = Ytest[:,0]
```

```
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(128, input_shape=(7,), activation='relu'),
  tf.keras.layers.Dense(1)
])

opt = tf.keras.optimizers.Adam(0.01)
model.compile(optimizer=opt, loss='mse')

r = model.fit(X, Y, validation_data=(Xtest, Ytest), epochs=1000)
```

```
plt.plot(r.history['loss'], label='loss')
```

```
Xpred = Xtest[35,:]
Xpred = Xpred.reshape(7,1).T
print(Xpred)
```

```
Ypred = model.predict(Xpred)
print(Ypred)
```

```
import matplotlib.pyplot as plt
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
```
